```python
from bs4 import BeautifulSoup
import requests

# URL target
url = 'https://sulut.bps.go.id/indicator/12/958/1/jumlah-penduduk-menurut-provinsi-di-indonesia.html'

# Mengirim permintaan GET dan mendapatkan konten HTML
response = requests.get(url)
html_content = response.content

# Membuat objek BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Mencari semua baris dalam tag tbody
rows = soup.select('tbody tr')

data_penduduk = []
for row in rows:
    province = row.select_one('td.vervar').get_text()
    year_2020 = row.select('td.text-right')[0].get_text()
    year_2021 = row.select('td.text-right')[1].get_text()
    year_2022 = row.select('td.text-right')[2].get_text()

    data_penduduk.append({
        'provinsi': province,
        '2020': year_2020,
        '2021': year_2021,
        '2022': year_2022
    })
```
